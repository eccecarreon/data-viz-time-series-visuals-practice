# Solution

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
url ="https://docs.google.com/spreadsheets/d/e/2PACX-1vT_jChgNsQbHbg4TGepzIqk8XC9DTIKmyyxb1upo5cfZCgbfIUQc2ZC0YMzuU5uApP140Ob49KBjdqh/pub?gid=1198589591&single=true&output=csv"
df = pd.read_csv(url)
df
```

```python
# 1) Convert the date column to datetime dtype.
# Double check if the results are correct and as expected --in the appropriate format! Correct if not.

df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
df['date']
```

```python
# This is the correct format!
```

```python
# 2) Set the date column as the index.
# What type is it?
df = df.set_index('date')
df
```

```python
# 3) Filter out only the required data:
# Keep only the data from year 2000 and later.
# Keep only the following features/ columns: 
# precipitation, mean_temp, min_temp, max_temp, snow_depth
df = df.loc['2000':, ['precipitation', 'mean_temp', 'min_temp', 'max_temp', 'snow_depth']]
df
```

```python
# 4) Impute any missing values.
# Check on each feature and think about which imputation method makes sense. It may not be the same method for every feature!
df.isna().sum()
```

```python
# Interpolate temperature measures
df['mean_temp'] = df['mean_temp'].interpolate()
df['max_temp'] = df['max_temp'].interpolate()
df['min_temp'] = df['min_temp'].interpolate()
```

```python
# Check null values
df.isna().sum()
```

```python
# Decided to fill the null values with '0' for precipitation and snow.
df['precipitation'] = df['precipitation'].fillna(0)
df['snow_depth'] = df['snow_depth'].fillna(0)
```

```python
# Check if the null values have been filled
df.isna().sum()
```

```python
# Import dataes as mdates from matplotlib
from matplotlib import dates as mdates
```

```python
# 5. Plot the mean, min, and max temps for 2018-2020 on one figure.

# Note: you may want to make a temporary DataFrame that you will use to answer this question only.

#     * Use a combination of pandas, matplotlib tick formatters, tick locators, and the fig.autofmt_xdate method to:
#         * Create a wide figure (e.g. figsize=(12,3))
#         * Set the title to "Daily Temperature"
#         * Set the ylabel to "Degrees"
#         * Customize the major xaxis ticks to:
#             * Have major xaxis ticks every 1 year.
#             * Use 4-digit years for the labels
#             * Rotate the major ticks' labels 90 degrees.
#         * Customize the minor xaxis ticks to:
#             * Have minor xaxis ticks every 3 months
#             * Use 3-letter month names for the labels.
```

```python
temp_cols = ['mean_temp','min_temp','max_temp']
plot_df = df.loc['2018':,temp_cols]
plot_df
```

```python
# Plot the data
ax = plot_df.plot(figsize=(12,5))
ax.set(title='Daily Temperature', ylabel='Degrees')

# Customize major ticks
major_loc = mdates.YearLocator(1)
major_fmt = mdates.DateFormatter("%Y")

ax.xaxis.set_major_locator(major_loc)
ax.xaxis.set_major_formatter(major_fmt)

# Customize minor ticks
minor_loc = mdates.MonthLocator(interval=3)
minor_fmt = mdates.DateFormatter("%b")

ax.xaxis.set_minor_locator(minor_loc)
ax.xaxis.set_minor_formatter(minor_fmt)

# Edit final xdates
fig = ax.get_figure()
fig.autofmt_xdate(which='major', rotation=90);
```

```python
# 6. Plot the precipitation and snow depth for 2000-2020

# Note: you may want to make a temporary DataFrame that you will use to answer this question only.

#     * Use a combination of pandas, matplotlib tick formatters, tick locators, and the fig.autofmt_xdate method to:
#         * Create a wide figure (e.g. figsize=(12,3))
#         * With the title of "Daily Precipitation"
#         * With a ylabel of "Inches"
#         * Customize the major xaxis ticks to:
#             * Have major xaxis ticks every 5 years.
#             * Use 4-digit years for the labels
#             * Rotate the major ticks' labels 45 degrees.
#         * Customize the minor xaxis ticks to:
#             * Have minor xaxis ticks every 3 months
#             * Do not label the minor ticks on the x-axis.
```

```python
precip_cols = ['precipitation', 'snow_depth']
plot_df = df.loc["2000":, precip_cols]
plot_df
```

```python
# Plot the data
ax = plot_df.plot(figsize=(12,5))
ax.set(title='Daily Precipitation and Snow Fall', ylabel='Inches')

# Customize major ticks
major_loc = mdates.YearLocator(5)
major_fmt = mdates.DateFormatter("%Y")

ax.xaxis.set_major_locator(major_loc)
ax.xaxis.set_major_formatter(major_fmt)

# Customize minor ticks
minor_loc = mdates.MonthLocator(interval=3)

ax.xaxis.set_minor_locator(minor_loc)

# Edit final xdates
fig = ax.get_figure()
fig.autofmt_xdate();
```
